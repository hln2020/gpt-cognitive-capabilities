```python
def get_response(prompt, model='text-davinci-003'):
    completions = ai.Completion.create(engine=model,
                                       temperature=0, 
                                       max_tokens=20,
                                       n=1,
                                       stop=None,
                                       prompt=prompt,
                                       logprobs=1,
                                      )
    
    return completions
```
